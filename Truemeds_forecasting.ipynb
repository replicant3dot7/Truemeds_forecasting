## Classifying the SKU time-series(demand) based on Average Demand Interval(ADI) and square of coefficiient of variation(CV^2)

```python


# defining time series classification function

def classify_demand_series(series):
    # Ensure the series is a numpy array
    series = np.array(series)
    
    # Calculate the Coefficient of Variation (CV)
    mean_demand = np.mean(series)
    std_demand = np.std(series)
    cv = std_demand / mean_demand if mean_demand != 0 else np.inf
    
    # Calculate the Average Inter-demand Interval (ADI)
    non_zero_demands = np.where(series != 0)[0]
    if len(non_zero_demands) > 1:
        adi = np.mean(np.diff(non_zero_demands))
    else:
        adi = np.inf
    
    # Classification rules
    if cv < 0.5 and adi < 1.32:
        classification = 'Smooth'
    elif cv >= 0.5 and adi < 1.32:
        classification = 'Erratic'
    elif cv < 0.5 and adi >= 1.32:
        classification = 'Intermittent'
    elif cv >= 0.5 and adi >= 1.32:
        classification = 'Lumpy'
    else:
        classification = 'Unclassified'
    
    return classification
```

```python
# classifying all the sku time-series
df_cat = df.copy()

# Extract product names
sku_names = df_cat['SKU Code']

# Extract time series data
time_series_data = df_cat.drop(columns=['SKU Code'])

# Classify each sku
classifications = {sku_names[i]: classify_demand_series(time_series_data.iloc[i]) for i in range(len(sku_names))}

# Converting to DataFrame for better visualization
classification_df = pd.DataFrame(list(classifications.items()), columns=['SKU Code', 'Classification'])
print(classification_df)
```

```python
from prophet import Prophet
from sklearn.metrics import mean_squared_error, mean_squared_log_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_absolute_percentage_error
import statistics
```

```python
# Using Prophet combing all the SKUs

df_prop = df.copy()

# Reshape the DataFrame for Prophet
data = {'ds': [], 'y': []}
for col in df_prop.columns[1:]:
    month_number = int(col.split('Month')[1])
    ds_values = [f'2024-{month_number:02d}-{i+1:02d}' for i in range(len(df_prop))]
    y_values = df_prop[col].values
    data['ds'].extend(ds_values)
    data['y'].extend(y_values)

# Convert to DataFrame
df_prophet = pd.DataFrame(data)

# Ensure 'ds' column is in datetime format
df_prophet['ds'] = pd.to_datetime(df_prophet['ds'], errors='coerce')

# Drop rows with missing dates
df_prophet.dropna(subset=['ds'], inplace=True)

# Initialize Prophet model
model = Prophet()

# Fit the model
model.fit(df_prophet)

# Make future dataframe for forecasting
future = model.make_future_dataframe(periods=12, freq='M')  # Assuming 12 months forecast

# Forecast
forecast = model.predict(future)

# Plot forecast
fig = model.plot(forecast)
```

**Combing all the SKU sales data and predicting on them is not the right approach as these SKUs are different from each other.** 
**So using Prophet to forecast on each SKU and using evaluation metrics like RMSE, MAE and MAPE to understand the performance of the model.**

```python
# Using Prophet for each SKU sales data

df_prop = df.copy()

# Initialize an empty DataFrame to store all forecasts
all_forecasts = pd.DataFrame()

# Iterate over each sku
for product in df_prop['SKU Code']:
    # Extract the sales data for the current product
    product_data = df_prop[df_prop['SKU Code'] == product].drop(columns=['SKU Code']).transpose()
    product_data.reset_index(inplace=True)
    product_data.columns = ['ds', 'y']
    
    # Extract month number from 'ds' and create proper datetime format
    product_data['ds'] = product_data['ds'].apply(lambda x: pd.Timestamp(f'2023-{x.split(" ")[1]}-{int(x.split(" ")[0][len("Month"):]) if x.split(" ")[0][len("Month"):] else 1:02d}'))

    # Drop rows with invalid month numbers
    product_data = product_data[product_data['ds'].notnull()]

    # Initialize Prophet model
    model = Prophet()

    # Fit the model
    model.fit(product_data)

    # Make future dataframe for forecasting
    future = model.make_future_dataframe(periods=2, freq='M')  # Assuming 2 months forecast

    # Forecast
    forecast = model.predict(future)

    # Add the forecast to the DataFrame
    forecast['SKU Code'] = product
    all_forecasts = pd.concat([all_forecasts, forecast], ignore_index=True)

# Plot forecasts for all products
fig = model.plot(all_forecasts)
```

**Splitting the data into training (80%-7 data points) and testing data (20%-2 data points)**

```python
# Assuming df contains the sales data with products in rows and months in columns

from sklearn.model_selection import train_test_split

df_pro=df.copy()
# Define the proportion of data to be used for training
train_size = 0.8  # 80% of data for training, 20% for testing

# Split the data for each product
train_data = pd.DataFrame()
test_data = pd.DataFrame()

for product in df.index:
    product_data = df_pro.loc[product]
    train_product, test_product = train_test_split(product_data, train_size=train_size, shuffle=False)
    train_data = pd.concat([train_data, train_product], axis=1)
    test_data = pd.concat([test_data, test_product], axis=1)

# Transpose the data to have products in columns and months in rows
df_train = train_data.T
skus=df_pro.iloc[:,0]
df_test = pd.concat([skus,test_data.T],axis=1)

# Check the dimensions of the training and testing sets
print("Training data dimensions:", train_data.shape)
print("Testing data dimensions:", test_data.shape)
```

## Using prophet model to forecast on training(Months 1-7) and testing data(Months 8,9). Also, predicting for the 10th month.
## Calculating evaluation metrics like RMSE, MAE and MAPE for the test data

```python
# Convert train and test data to DataFrames
train_df_prop = df_train.copy()
test_df_prop = df_test.copy()

# Initialize an empty list to store evaluation metrics for each SKU
evaluation_metrics_prop = []

# Iterate over each SKU and its corresponding train/test data
for sku in train_df_prop['SKU Code']:
    try:
        # Prepare the DataFrame for Prophet
        train_sku_prop = train_df_prop[train_df_prop['SKU Code'] == sku].drop(columns=['SKU Code'])
        train_sku_prop = train_sku_prop.transpose().reset_index()
        train_sku_prop.columns = ['ds', 'y']
        train_sku_prop['ds'] = pd.date_range(start='2022-01-01', periods=len(train_sku_prop), freq='M')

        # Initialize Prophet model
        model = Prophet()

        # Fit the model on training data
        model.fit(train_sku_prop)

        # Prepare the future DataFrame for forecasting
        future = pd.DataFrame({'ds': pd.date_range(start='2022-08-01', periods=3, freq='M')})  # Forecast for months 8, 9, and 10

        # Make predictions for the test period and the 10th month
        forecast = model.predict(future)

        # Extract predicted values for months 8 and 9(test data)
        y_pred = forecast['yhat'].values[:2]

        # Extract actual values from the test data
        test_sku = test_df_prop[test_df_prop['SKU Code'] == sku].drop(columns=['SKU Code']).transpose().reset_index(drop=True)
        y_true = test_sku.values.flatten()
        print('y_true', y_true)
        print('y_pred', y_pred)

        # Calculate evaluation metrics for months 8 and 9(test data)
        rmse = np.sqrt(mean_squared_error(y_true, y_pred))
        mae = mean_absolute_error(y_true, y_pred)
        mape = np.mean(np.abs((y_true - y_pred) / y_true)) * 100  # Calculate MAPE

        # Get prediction for month 10
        month_10_forecast = forecast['yhat'].values[2]

        # Store evaluation metrics and the 10th month prediction for the current SKU
        evaluation_metrics_prop.append({
            'SKU Code': sku,
            'RMSE': rmse,
            'MAE': mae,
            'MAPE': mape,
            'Month 10 Forecast': month_10_forecast
        })

    except Exception as e:
        print(f"Error processing SKU {sku}: {e}")

# Convert evaluation metrics to DataFrame for easier analysis
evaluation_df_prop = pd.DataFrame(evaluation_metrics_prop)

# Print or analyze evaluation metrics as needed
print(evaluation_df_prop)
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error
import keras
from keras.layers import Dense, LSTM
from keras.models import Sequential
from tensorflow.keras.optimizers import Adam
from keras.callbacks import EarlyStopping
from sklearn.model_selection import train_test_split
```

```python

# Function to prepare data for LSTM
def prepare_lstm_data(data, n_steps):
    X, y = [], []
    for i in range(len(data)):
        end_ix = i + n_steps
        if end_ix > len(data)-1:
            break
        seq_x, seq_y = data[i:end_ix], data[end_ix]
        X.append(seq_x)
        y.append(seq_y)
    return np.array(X), np.array(y)

# Normalize the data
scaler = MinMaxScaler()

# Convert train and test data to DataFrames
train_df = df_train.copy()
test_df = df_test.copy()

# Initialize an empty list to store evaluation metrics for each SKU
evaluation_metrics = []

# Number of steps for LSTM
n_steps = 7  # Last 7 months to predict the next month

# Iterate over each SKU and its corresponding train/test data
for sku in train_df['SKU Code']:
    try:
        # Prepare the training data
        train_sku = train_df[train_df['SKU Code'] == sku].drop(columns=['SKU Code']).values.flatten()
        train_sku_scaled = scaler.fit_transform(train_sku.reshape(-1, 1)).flatten()
        print(train_sku)
        print(train_sku_scaled)

        # Prepare test data
        test_sku = test_df[test_df['SKU Code'] == sku].drop(columns=['SKU Code']).values.flatten()
        test_sku_scaled = scaler.transform(test_sku.reshape(-1, 1)).flatten()
        print(test_sku)
        print(test_sku_scaled)

        # Prepare the LSTM training data
        X_train, y_train = prepare_lstm_data(train_sku_scaled, n_steps)

        # Reshape data into 3D format [samples, timesteps, features]
        X_train = X_train.reshape((X_train.shape[0], X_train.shape[1], 1))

        # Define LSTM model
        model = Sequential()
        model.add(LSTM(50, activation='relu', input_shape=(n_steps, 1)))
        model.add(Dense(1))
        model.compile(optimizer='adam', loss='mse')

        # Fit the model
        model.fit(X_train, y_train, epochs=200, verbose=0)

        # Prepare the input data for prediction (last 7 months from the train data)
        last_train_data = train_sku_scaled[-n_steps:]
        last_train_data = last_train_data.reshape((1, n_steps, 1))

        # Predict month 8, month 9, and month 10
        predictions = []
        for _ in range(3):
            yhat = model.predict(last_train_data, verbose=0)
            predictions.append(yhat[0, 0])
            last_train_data = np.append(last_train_data[:, 1:, :], yhat.reshape((1, 1, 1)), axis=1)

        # Rescale predictions back to original scale
        y_pred = scaler.inverse_transform(np.array(predictions).reshape(-1, 1)).flatten()

        # Extract actual values from the test data for months 8 and 9
        y_true = test_sku.flatten()

        # Calculate evaluation metrics for months 8 and 9(test data)
        rmse = np.sqrt(mean_squared_error(y_true, y_pred[:2]))
        mae = mean_absolute_error(y_true, y_pred[:2])
        mape = np.mean(np.abs((y_true - y_pred[:2]) / y_true)) * 100  # Calculate MAPE

        # Store evaluation metrics for the test data and the 10th month prediction for the current SKU
        evaluation_metrics.append({
            'SKU Code': sku,
            'RMSE': rmse,
            'MAE': mae,
            'MAPE': mape,
            'Month 10 Prediction': y_pred[2]
        })

    except Exception as e:
        print(f"Error processing SKU {sku}: {e}")

# Convert evaluation metrics to DataFrame for easier analysis
evaluation_df = pd.DataFrame(evaluation_metrics)

# Print evaluation metrics
print(evaluation_df)
```

```python

# Function to prepare data for LSTM
def prepare_lstm_data(data, n_steps):
    X, y = [], []
    for i in range(len(data)):
        end_ix = i + n_steps
        if end_ix > len(data)-1:
            break
        seq_x, seq_y = data[i:end_ix], data[end_ix]
        X.append(seq_x)
        y.append(seq_y)
    return np.array(X), np.array(y)

# Normalize the data
scaler = MinMaxScaler()

# Convert train and test data to DataFrames
train_df = df_train.copy()
test_df = df_test.copy()

# Initialize an empty list to store evaluation metrics for each SKU
evaluation_metrics = []

# Number of steps for LSTM
n_steps = 5  # Last 5 months to predict the next month

# Iterate over each SKU and its corresponding train/test data
for sku in train_df['SKU Code']:
    try:
        # Prepare the training data
        train_sku = train_df[train_df['SKU Code'] == sku].drop(columns=['SKU Code']).values.flatten()
        train_sku_scaled = scaler.fit_transform(train_sku.reshape(-1, 1)).flatten()
        print(f'Training data for SKU {sku}:', train_sku)
        print(f'Scaled training data for SKU {sku}:', train_sku_scaled)

        # Prepare test data
        test_sku = test_df[test_df['SKU Code'] == sku].drop(columns=['SKU Code']).values.flatten()
        test_sku_scaled = scaler.transform(test_sku.reshape(-1, 1)).flatten()
        print(f'Test data for SKU {sku}:', test_sku)
        print(f'Scaled test data for SKU {sku}:', test_sku_scaled)

        # Prepare the LSTM training data
        X_train, y_train = prepare_lstm_data(train_sku_scaled, n_steps)

        # Debugging the shapes
        print(f'Shape of X_train for SKU {sku}:', X_train.shape)
        print(f'Shape of y_train for SKU {sku}:', y_train.shape)

        # Check if there are enough samples for training
        if X_train.shape[0] == 0 or y_train.shape[0] == 0:
            raise ValueError(f"Not enough data to create training samples for SKU {sku}.")

        # Reshape data into 3D format [samples, timesteps, features]
        X_train = X_train.reshape((X_train.shape[0], X_train.shape[1], 1))

        # Define LSTM model
        model = Sequential()
        model.add(LSTM(50, activation='relu', input_shape=(n_steps, 1)))
        model.add(Dense(1))
        model.compile(optimizer='adam', loss='mse')

        # Fit the model
        model.fit(X_train, y_train, epochs=200, verbose=0)

        # Prepare the input data for prediction (last 5 months from the train data)
        last_train_data = train_sku_scaled[-n_steps:]
        last_train_data = last_train_data.reshape((1, n_steps, 1))

        # Debugging the last_train_data
        print(f'Last training data for prediction for SKU {sku}:', last_train_data)
        print(f'Shape of last_train_data for prediction for SKU {sku}:', last_train_data.shape)

        # Predict month 8, month 9, and month 10
        predictions = []
        for _ in range(3):
            yhat = model.predict(last_train_data, verbose=0)
            predictions.append(yhat[0, 0])
            last_train_data = np.append(last_train_data[:, 1:, :], yhat.reshape((1, 1, 1)), axis=1)

        # Rescale predictions back to original scale
        y_pred = scaler.inverse_transform(np.array(predictions).reshape(-1, 1)).flatten()

        # Extract actual values from the test data for months 8 and 9
        y_true = test_sku.flatten()

        print(f"SKU: {sku}")
        print(f"Actual values (y_true): {y_true}")
        print(f"Predicted values (y_pred): {y_pred}")

        # Ensure y_pred has at least 2 elements
        if len(y_pred) < 2:
            raise ValueError(f"Not enough predicted values for SKU {sku}. Needed 2, got {len(y_pred)}")

        # Calculate evaluation metrics for months 8 and 9 (test data)
        rmse = np.sqrt(mean_squared_error(y_true, y_pred[:2]))
        mae = mean_absolute_error(y_true, y_pred[:2])
        mape = np.mean(np.abs((y_true - y_pred[:2]) / y_true)) * 100  # Calculate MAPE

        # Store evaluation metrics for the test data and the 10th month prediction for the current SKU
        evaluation_metrics.append({
            'SKU Code': sku,
            'RMSE': rmse,
            'MAE': mae,
            'MAPE': mape,
            'Month 10 Prediction': y_pred[2] if len(y_pred) > 2 else None
        })

    except Exception as e:
        print(f"Error processing SKU {sku}: {e}")

# Convert evaluation metrics to DataFrame for easier analysis
evaluation_df = pd.DataFrame(evaluation_metrics)

# Print evaluation metrics
print(evaluation_df)
```
